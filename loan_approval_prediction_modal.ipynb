```python
df['assets']=df.residential_assets_value+df.commercial_assets_value+df.luxury_assets_value+df.bank_asset_value
```

```python
df.drop(columns=['residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'],inplace=True)
```

```python
df
```

```python
df.education.unique()
```

```python
def clean_text(st):
  st=st.strip()
  return st
```

```python
df['education']=df['education'].apply(clean_text)
```

```python
df.education.unique()
```

```python
df['education']=df['education'].replace(['Graduate', 'Not Graduate'],[1,0])
```

```python
df
```

```python
df['self_employed'].unique()
```

```python
df['self_employed']=df['self_employed'].apply(clean_text)
```

```python
df['self_employed'].unique()
```

```python
df['self_employed']=df['self_employed'].replace(['Yes', 'No'],[1,0])
```

```python
df
```

```python
df['loan_status'].unique()
```

```python
df['loan_status']=df['loan_status'].apply(clean_text)
```

```python
df['loan_status'].unique()
```

```python
df['loan_status']=df['loan_status'].replace(['Approved', 'Rejected'],[1,0])
```

```python
df
```

```python
df.columns
```

```python
x=df.drop(columns=['loan_status'])
y=df['loan_status']
```

```python
y
```

Train Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
score
```
